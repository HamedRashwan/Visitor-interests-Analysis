```python
def extract_browser(user_agent):
    ua = parse(user_agent)
    return ua.browser.family

def extract_os(user_agent):
    ua = parse(user_agent)
    return ua.os.family

def extract_device(user_agent):
    ua = parse(user_agent)
    return ua.device.family
```

```python
df['Browser'] = df['UserAgent'].apply(extract_browser)
df['OS'] = df['UserAgent'].apply(extract_os)
df['Device'] = df['UserAgent'].apply(extract_device)
```

```python
df
```

```python
df.drop(columns=['UserAgent'], inplace=True)
```

```python
df
```

```python
def get_location(ip_address):
    g = geocoder.ip(ip_address)
    return g.latlng

df['Latitude'] = df['IP'].apply(lambda x: get_location(x)[0] if get_location(x) else None)
df['Longitude'] = df['IP'].apply(lambda x: get_location(x)[1] if get_location(x) else None)
```
